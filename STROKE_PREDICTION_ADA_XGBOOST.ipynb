# DATA CLEANING AND PRE PROCESSING (FEATURE ENGINEERING)

From the analysis of graphs above, we see a number of missing values, skews, kurtosis and so on which can impact the model performance significantly. Hence we will need to handle them.


```python
# Check for missing values
data = df.copy()
df.isnull().sum()
```

```python
# Check for Duplicated values

duplicate_rows = data.duplicated()
duplicate_rows.sum()
```

```python
#To preserve relation between continuous variables like BMI, we can use KNN imputation to impute missing values
from sklearn.neighbors import KNeighborsRegressor

def knn_impute(df, target_column):
    numeric_df = df.select_dtypes(include=[np.number])

    complete_columns = numeric_df.loc[:, numeric_df.isna().sum() == 0].columns

    # Split data into training and test data
    mask_non_missing = numeric_df[target_column].isna() == False
    mask_missing = numeric_df[target_column].isna() == True

    # Prepare training and test data
    X_train = numeric_df.loc[mask_non_missing, complete_columns]
    y_train = numeric_df.loc[mask_non_missing, target_column]

    X_test = numeric_df.loc[mask_missing, complete_columns]

    # Train and predict using KNN
    knn = KNeighborsRegressor()
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    # Fill missing values with predictions
    df.loc[mask_missing, target_column] = y_pred

    return df

# Find columns with missing values
columns_with_missing = [col for col in data.columns if data[col].isnull().sum() != 0]

# Apply KNN imputation to each column with missing values
for column in columns_with_missing:
    data = knn_impute(data, column)
```

```python

data.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data_original = data.copy()
# Dropping "id"
data_original = data_original.drop("id",axis=1)

# Encoding Categorical Columns
cols_to_be_encoded = ['gender', 'ever_married', 'work_type',
                   'Residence_type', 'smoking_status']

# Applying OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

encoded_data = encoder.fit_transform(data_original[cols_to_be_encoded])

#Create DataFrame with encoded column features
encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(cols_to_be_encoded)
)

encoded_data = pd.concat([
    data_original.drop(columns=cols_to_be_encoded),
    encoded_df
], axis=1)
```

```python
encoded_data
```

```python
#Handling outliers using Z-score method

def detect_outliers_zscore(encoded_data, columns: list, threshold: float = 3) -> dict:

    def count_zscore_outliers(series: pd.Series) -> int:
        # Calculate z-scores
        z_scores = (series - series.mean()) / series.std()

        # Count values beyond threshold
        return (abs(z_scores) > threshold).sum()

    return {col: count_zscore_outliers(df[col]) for col in columns}

numerical_columns = ['age', 'avg_glucose_level', 'bmi']
outliers = detect_outliers_zscore(data, numerical_columns)
outliers
```

```python

fig, axes = plt.subplots(3, 2, figsize=(15, 15))


for i, var in enumerate(numerical_columns):
    # Histogram
    sns.histplot(data=encoded_data, x=var, ax=axes[i, 0], kde=True)
    axes[i, 0].set_title(f'Histogram of {var} (Transformed)')

    # Box plot
    sns.boxplot(data=encoded_data, y=var, ax=axes[i, 1])  # Removed x='stroke' as it's not in transformed data
    axes[i, 1].set_title(f'Box Plot of {var} (Transformed)')

plt.tight_layout()
plt.show()
```

Since we have observed a lot of imbalanced data in our dataset, To improve performance there is a strong need to balance them. Here we will be using **Adaptive Synthetic Sampling**.


This is a technique that generates synthetic data to improve classification accuracy for underrepresented classes in imbalanced datasets. ADASYN is an enhanced version of the **Synthetic Minority Over-sampling Technique**

```python
#USing ADASYN to improve accuracy
from imblearn.over_sampling import ADASYN

X = encoded_data.drop('stroke', axis=1)
y = encoded_data['stroke']

# Apply ADASYN
adasyn = ADASYN(random_state=42)
X_resampled, y_resampled = adasyn.fit_resample(X, y)

# Create the final balanced Data Frame
cleaned_data = pd.DataFrame(X_resampled, columns=X.columns)
cleaned_data['stroke'] = y_resampled

print(cleaned_data['stroke'].value_counts())
```

# TRAINING THE MODEL WITH XGBOOST

After a bit of tuning, these parameters gave me the most optimal result with best generalization:



*   **colsample_bytree**: 0.5. This parameter determines the fraction of columns to be randomly sampled for each tree. By setting it to 0.5, I use 50% of the columns for each tree

*   **gamma**: 0.2. Gamma serves as a parameter for the tree-based learners and controls whether a given node will split based on the expected reduction in the loss after the split. Choosing a higher gamma value makes the algorithm more conservative, leading to fewer splits and smaller trees as it becomes cautious about splitting

*   **learning_rate**: 0.1. This refers to the step size shrinkage utilized in the update process to prevent overfitting

*   **max_depth**: 7. This parameter defines the maximum depth of a tree and is employed to control overfitting. A larger depth implies a more complex model and higher chances of overfitting


*   **subsample**: 0.5. This parameter determines the fraction of observations to be randomly sampled for each tree. By setting it to 0.5, I use 50% of the data for each tree.







```python

from sklearn.model_selection import train_test_split
import xgboost as xgb

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    cleaned_data.drop('stroke', axis=1),
    cleaned_data['stroke'],
    test_size=0.2,
    random_state=42
)

# Initialize XGBoost classifier with optimized hyperparameters
model = xgb.XGBClassifier(
    colsample_bytree=0.5,    # Fraction of features used for each tree
    gamma=0.2,               # Minimum loss reduction for split
    learning_rate=0.1,      # Slower learning for better generalization
    max_depth=7,             # Maximum tree depth
    min_child_weight=5,      # Minimum sum of instance weights in child
    subsample=0.5,           # Fraction of samples used for each tree
    eval_metric='logloss',   # Loss function for evaluation
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Generating Report and Analysis of result
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```

The number of false negatives and false positives is quite low for our model

```python
# ROC Curve
y_prob = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

 The AUC value is 0.99 which is nearly perfect. The AUC value is an aggregate measure of the model's performance across all classification thresholds.

```python
# Plotting learning curve
train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training accuracy')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, color='green', linestyle='--', marker='s', markersize=5, label='Validation accuracy')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend(loc='lower right')
plt.show()
```

We observe a good learning curve that shows healthy model learning behavior. Since the curves remain together, there seems to be **no overfitting** issues. Since the accuracy is high, **no underfitting** is observed. A clear convergence pattern is also observed

# INFERENCE
 From the ROC and Learning curve graphs, we see that the XG-Boost model that has been tuned and modelled is fairly robust and generalizes well to unseen data. The relatively low number of false positives and false negatives in the confusion matrix further supports our claim.

 Comparing other models such as Logistic Regression and SVM. The results show that XGBoost outperformed these models with significantly higher ROC AUC and overall classification metrics.

 Some of the main reasons why XG-Boost performs best here is because:


1.   **Handling Non-linear Relationships:** XGBoost, being a tree-based ensemble method, excels at capturing complex non-linear relationships between features and the target variable (stroke).
2.   **Feature Interactions:** XGBoost can automatically identify and leverage interactions between different features. This is crucial in stroke prediction, where the risk is often influenced by a combination of factors ( age, hypertension, glucose levels) rather than individual features in isolation. Linear models like Logistic Regression struggle to capture such interactions without explicit feature engineering.
3.  **Regularization and Overfitting Control:** XGBoost incorporates regularization techniques that help prevent overfitting. This is essential when dealing with limited datasets, as it ensures the model generalizes well to unseen data. While SVM also has regularization, XGBoost's ensemble nature provides further robustness against overfitting.
4. The specific characteristics of the stroke dataset favors XGBoost. The presence of mixed data types (numerical and categorical) and potential outliers can be handled effectively by XGBoost's tree-based structure.
5.  **Flexibility and Optimization**: XGBoost offers a wide range of hyperparameters that can be tuned to optimize the model for specific datasets and tasks.



 The Extreme Gradient Boost model, seems like a promising model for stroke prediction. By identifying and prioritizing important features, this model can assist medical practitioners in making informed decisions about stroke prevention and early interventions.



# Comparing Other Models

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize other classifiers for comparison
classifiers = {

    "RandomForest": RandomForestClassifier(random_state=42),
    "LogisticRegression": LogisticRegression(random_state=42, max_iter=1000),
    "SVM": SVC(random_state=42, probability=True),
    "DecisionTree": DecisionTreeClassifier(random_state=42)
}


results = {}
for name, model in classifiers.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"--- {name} ---")
    print(f"{name} Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))


    #ROC Curve
    try:
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve {name}(area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Receiver Operating Characteristic (ROC) for {name}')
        plt.legend(loc="lower right")
        plt.show()
    except AttributeError:
        print(f"ROC curve not available for {name}")


# Compare results (example: print accuracy)
print("\nModel Comparison (Accuracy):")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.4f}")
```

From the above results, we can see that the performance of XGboost model is higher than models like regression and SVM in terms of ROC and AUC.


 XGBoost's tree-based structure excels at capturing complex patterns in data, unlike linear models.  This is important because stroke risk likely involves intricate interactions of different factors. Linear models require manual feature engineering to capture these interactions, which can be tedious and error-prone.
